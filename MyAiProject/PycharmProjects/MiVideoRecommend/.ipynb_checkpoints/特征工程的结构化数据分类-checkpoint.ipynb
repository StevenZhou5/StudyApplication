```python
def df_to_datasets(dataframe,shuffle=True,batch_size=32):
    dataframe = dataframe.copy()
    labels = dataframe.pop('target') # 将标签列取出,取出后dataframe中将没有'target'列了
    ds = tf.data.Dataset.from_tensor_slices((dict(dataframe),labels)) #ds是拆分成了303个字典组成的组合
    if shuffle:
        ds.shuffle(buffer_size=len(dataframe)) #shuffle的功能为打乱dataset中的元素，它有一个参数buffersize，表示打乱时使用的buffer的大小：
    ds = ds.batch(batch_size) # 按照每份batch_size个，分成（总数/batch_size）个组
    return ds
dataframe = dataframe.copy()
labels = dataframe.pop('target') # 将标签列取出
```
