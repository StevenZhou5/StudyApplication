```python
import numpy as np
import pandas as pd
import tensorflow as tf 
import matplotlib.pyplot as plt
from tensorflow import keras
from tensorflow.keras import layers

from sklearn.model_selection import train_test_split


# 注意每次Embedding的参数都不一样，所以必须要把模型存储起来

model = keras.Sequential()
# 1000代表的就是数字的最大上限（原维数），64就是新的维，input_length就是原先的数字最是有多少个1000以内的数字表示的（相当于电影特征就多有20个，每个标签对应一个数字）
model.add(layers.Embedding(1000, 64, input_length=10)) 
# the model will take as input an integer matrix of size (batch,
# input_length).
# the largest integer (i.e. word index) in the input should be no larger
# than 999 (vocabulary size).
# now model.output_shape == (None, 10, 64), where None is the batch
# dimension.

input_array = np.random.randint(1000, size=(5, 10))

model.compile('rmsprop', 'mse')
output_array = model.predict(input_array)
assert output_array.shape == (5, 10, 64)
```
