```python
import numpy as np
import torch
```

```python
# 热身：先用Numpy来实现两层的神经网络
# 一个全连接用ReLU激活，一个隐藏层，没有bias，
# 

def RELU(Z):
    return np.maximum(Z,0)

def DZRELU(A):
    return A >0

# 第一步定义初始化数据和参数
Learning_rate = 1e-5
M = 1000 # 1000组训练数据
NL0,NL1,NL2 = 5,10,1 # 第0层64和神经元，第一层100个神经元，最后一层  一个神经元
train_x = np.random.randn(NL0,M)
train_y = np.random.randn(1,M)

W1 = np.random.randn(NL1,NL0)
B1 = np.random.randn(NL1,1)
W2 = np.random.randn(NL2,NL1)
B2 = np.random.randn(NL2,1)


print(train_x.shape)
print(train_y.shape)
print(W1.shape)
print(B1.shape)
print(W2.shape)

assert W1.shape[1] == train_x.shape[0]

for t in range(500):
    # 第二步进行前项传播
    A0 = train_x
    Z1 = np.dot(W1,A0) +B1
    A1 = RELU(Z1)
    Z2 = np.dot(W2,A1) + B2
    A2 = Z2 # 线性激活函数就等于Z2
#     print(A2.shape)

    # 计算输出一下Loss
    loss = (np.square(A2 - train_y).sum())/(2*M)
    print(t,":",loss)
    
    # 第三步进行反向传播：
    DY = 2.0*(A2 - train_y) #根据均方根Loss函数来计算DY
    DZ2 = DA2 = DY
    DW2 = np.dot(DZ2,A1.T)
    DB2 = DZ2

    DA1 = np.dot(W2.T,DZ2)
    DZ1 = DA1*DZRELU(A1)
    DW1 = np.dot(DZ1,A0.T)
    DB1 = DZ1

    # 第四步，执行梯度下降
    W1 = W1 - Learning_rate*DW1
    B1 = B1 - Learning_rate*DB1
    W2 = W2 - Learning_rate*DW2
    B2 = B2 - Learning_rate*DB2


```

```python








#怎么用 torch实现全连接神经网络
#1：跟踪Tnsor上的所有操作：设置属性 requires_grad = True
#2: 自动计算所有梯度：调用 .backward()
#3: 停止跟踪Tensor：方式1：调用.detach()
#                 方式2：代码块'with torch.no grad'
#4: 如果Tensor不仅仅是表里，则需要指定gradient参数来指定张量的形式
```

```python
# 把numpy 变为torch
# 第一步，训练数据初始化
N,D_in,H,D_out = 64,1000,100,10 # 训练数据个数，输入层神经元个数，隐藏层神经元个数，输出层神经元个数
x = torch.randn(N,D_in) # N组训练数，输入层有D_in个神经元
y = torch.randn(N,D_out)

print(x.size()) # 查看x的shape

w1 = torch.randn(D_in,H) # 和TensorFlow一样是顺序定义的
w2 = torch.randn(H,D_out)

learning_rate = 1e-6
for it in range(500):
    # Forward前向传播
    h = x.mm(w1) # 矩阵的点乘法，用来求Z，这里没有常量偏移bias ，最终的sizey应该是N*H
    h_relu = h.clamp(min=0) # Relu激活函数，就是让小于0的值都变为0
    y_pred = h_relu.mm(w2) # 线性回归计算出预测值
    
    # compute loss 计算代价函数值
    loss = (y_pred - y).pow(2).sum().item() # 均方差loss，可以除以2N
    print("第",it,"次时的代价值为：",loss)
    
    # Backward pass反向传播
    grad_y_pred = 2.0 * (y_pred - y)
    grad_w2 = h_relu.t().mm(grad_y_pred) # 用h_relu和grad_y_pred来计算dw2 ,size最终为(H,D_dout)
    grad_h_relu = grad_y_pred.mm(w2.t()) #用grad_y_pred和w2来计算grad_h_relu,size最终为(N,H)
    grad_h = grad_h_relu.clone() # 用dA1*Relu激活函数对Z求导：正数为1，负数为0，最终size为(N,H)
    grad_h[h<0] = 0 # 把Z中小于0的位置的dZ的值置为0
#     print(grad_h)
    grad_w1 = x.t().mm(grad_h) # 最后在利用x和dZ来计算dW1，最终size为(D_in,H)
    
    # update weights of w1 and w2 更新w1和w2
    w1 -= learning_rate*grad_w1
    w2 -= learning_rate*grad_w2
    
```

```python
#自动计算所有梯度：调用 .backward()
x = torch.tensor(1.,requires_grad=True)
w = torch.tensor(2.,requires_grad=True)
b = torch.tensor(3.,requires_grad=True)

y = w*x +b
y.backward(retain_graph=True)  #自动计算所有梯度：调用 .backward()

# dy/dw =x
print(w)
w = w.clone() - w.grad
print(w.grad)
print(w)
# 同理 dy/dx = w
print (x.grad)
# 同理dy/db = 1 
print (b.grad)
```

```python
# 利用自动求导
N,D_in,H,D_out = 64,1000,100,10 # 训练数据个数，输入层神经元个数，隐藏层神经元个数，输出层神经元个数
x = torch.randn(N,D_in) # N组训练数，输入层有D_in个神经元
y = torch.randn(N,D_out)

print(x.size()) # 查看x的shape

w1 = torch.randn(D_in,H,requires_grad=True) # 这里必须声明requires_grad=True
w2 = torch.randn(H,D_out,requires_grad=True)# 这里必须声明requires_grad=True

learning_rate = 1e-6
for it in range(500):
    # Forward前向传播 ，只需要计算出预测函数的值
    y_pred = x.mm(w1).clamp(min=0).mm(w2) # 线性回归计算出预测值
    
    # compute loss 计算代价函数值
    loss = (y_pred - y).pow(2).sum() # 计算图
    print("第",it,"次时的代价值为：",loss.item())
    
    # Backward pass反向传播 :只需要调用待见函数的loss.backward()
    loss.backward()
    
    with torch.no_grad(): # 为了不要让计算图占内存，需要加上这句话
        w1 = w1 - learning_rate * w1.grad # 现在的版本会自动把w1.grad清除
        w1.requires_grad_(True) # 必须重置这个标志位
        w2 = w2 - learning_rate*w2.grad
        w2.requires_grad_(True) # 必须重置这个标志位

#     w1 = w1 - learning_rate*w1.grad
#     w2 = w2 - learning_rate*w2.grad
#         w1.grad.zero_() # 把w1的grad清零，不然每次会累加，不过试了发现好像并没有
#         w2.grad.zero_() # 把w2的grad清零，不然每次会累加，
```

```python
import torch.nn as nn # pytorch中的神经网络层

# 利用nn库进行精简
N,D_in,H,D_out = 64,1000,100,10 # 训练数据个数，输入层神经元个数，隐藏层神经元个数，输出层神经元个数
x = torch.randn(N,D_in) # N组训练数，输入层有D_in个神经元
y = torch.randn(N,D_out)

model = torch.nn.Sequential(
    torch.nn.Linear(D_in,H,bias=False), # 类似于keras的Dense全连接层
    torch.nn.ReLU(), # Relu激活层
    torch.nn.Linear(H,D_out,bias=False) # 全连接输出层
)

# 归一化初始权重
torch.nn.init.normal_(model[0].weight) # 第0层归一化
torch.nn.init.normal_(model[2].weight) # 输出层第二层归一化


# model = model.cuda # 如果要在GPU上运行的话

loss_fn = nn.MSELoss(reduction='sum')

learning_rate = 1e-6
for it in range(500):
    # Forward前向传播 
    y_pred = model(x)
    
    # compute loss 计算代价函数值
    loss = loss_fn(y_pred,y) # 计算图
    print("第",it,"次时的代价值为：",loss.item())
    
    
    # Backward pass反向传播 :只需要调用待见函数的loss.backward()
    loss.backward()
    
    with torch.no_grad(): # 为了不要让计算图占内存，需要加上这句话
        for param in model.parameters(): # 通过这种方式可以取到所有的param
            param -= learning_rate*param.grad # 对这些param做梯度下降
            
    model.zero_grad() # 这里可以吧grad清零，要在下一次训练前把之前的梯度进行调整
```

```python
import torch.nn as nn # pytorch中的神经网络层

# 使用Adam优化器

N,D_in,H,D_out = 64,1000,100,10 # 训练数据个数，输入层神经元个数，隐藏层神经元个数，输出层神经元个数
x = torch.randn(N,D_in) # N组训练数，输入层有D_in个神经元
y = torch.randn(N,D_out)

model = torch.nn.Sequential(
    torch.nn.Linear(D_in,H,bias=False), # 类似于keras的Dense全连接层
    torch.nn.ReLU(), # Relu激活层
    torch.nn.Linear(H,D_out,bias=False) # 全连接输出层
)

# 归一化初始权重
torch.nn.init.normal_(model[0].weight) # 第0层归一化
torch.nn.init.normal_(model[2].weight) # 输出层第二层归一化


# model = model.cuda # 如果要在GPU上运行的话

loss_fn = nn.MSELoss(reduction='sum')

learning_rate = 1e-3
optiimizer = torch.optim.Adam(model.parameters(),lr=learning_rate)

for it in range(500):
    # Forward前向传播 
    y_pred = model(x)
    
    # compute loss 计算代价函数值
    loss = loss_fn(y_pred,y) # 计算图
    print("第",it,"次时的代价值为：",loss.item())
    
    #在反向传播之前把grad清空
    optiimizer.zero_grad()
    
    # Backward pass反向传播 :只需要调用待见函数的loss.backward()
    loss.backward()
    
    # 利用optiimizer来进行梯度下降
    optiimizer.step()
            
```

```python
import torch.nn as nn # pytorch中的神经网络层

# 使用Adam优化器

N,D_in,H,D_out = 64,1000,100,10 # 训练数据个数，输入层神经元个数，隐藏层神经元个数，输出层神经元个数
x = torch.randn(N,D_in) # N组训练数，输入层有D_in个神经元
y = torch.randn(N,D_out)

# 自定义模型
class TwoLayerNet(torch.nn.Module):
    def __init__(self,D_in,H,D_out):
        super(TwoLayerNet,self).__init__()
        self.linear1 = torch.nn.Linear(D_in,H,bias=False)
        self.linear2 = torch.nn.Linear(H,D_out,bias=False)
    
    # 定义前向传播方法
    def forward(self, x):
        y_pred = self.linear2(self.linear1(x).clamp(min=0))#clamp相当于Relu的激活
        return y_pred
        

model = TwoLayerNet(D_in,H,D_out)

loss_fn = nn.MSELoss(reduction='sum')

learning_rate = 1e-3
optiimizer = torch.optim.Adam(model.parameters(),lr=learning_rate)

for it in range(500):
    # Forward前向传播 
    y_pred = model(x) # 这里相当于会调用forward(self, x)函数
    
    # compute loss 计算代价函数值
    loss = loss_fn(y_pred,y) # 计算图
    print("第",it,"次时的代价值为：",loss.item())
    
    #在反向传播之前把grad清空
    optiimizer.zero_grad()
    
    # Backward pass反向传播 :只需要调用待见函数的loss.backward()
    loss.backward()
    
    # 利用optiimizer来进行梯度下降
    optiimizer.step()
```
